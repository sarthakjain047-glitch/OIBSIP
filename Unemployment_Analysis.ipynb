```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("Downloads/Unemployment in India.csv")
df.head()
```

```python
# Check column names
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure()
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'])
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.title("Unemployment Rate Trend in India")
plt.show()
```

```python
state_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values()

plt.figure(figsize=(10,6))
state_avg.plot(kind='barh')
plt.xlabel("Average Unemployment Rate (%)")
plt.ylabel("State")
plt.title("State-wise Average Unemployment Rate in India")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Estimated Unemployment Rate (%)',
                'Estimated Employed',
                'Estimated Labour Participation Rate (%)']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Conclusion:")
print("This analysis shows unemployment trends in India over time.")
print("Certain states have consistently higher unemployment rates.")
print("The COVID period shows a sharp rise in unemployment.")
print("Visualizations help understand regional and temporal patterns clearly.")
```
